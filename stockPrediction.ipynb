```python
#Importing classes
import pandas as panda
import matplotlib
import numpy as numpy
from matplotlib import pyplot as plot

#Reading the NAS csv file to the dataframe
df = panda.read_csv("NAS.csv")

#Visualisation of the dataframe
df
```

```python
#Setting index as date
df['Date'] = panda.to_datetime(df.Date, format = '%Y-%m-%d')
df.index = df['Date']

#Plotting the stock market history
plot.figure(figsize=(16,8))
plot.plot(df['Close'], label = 'Close price history')
```

```python
#Sorting
data = df.sort_index(ascending = True, axis = 0)

#creating a separate dataset
new_data = panda.DataFrame(index = range(0, len(df)), columns = ['Date', 'Close'])

for x in range(0, len(data)):
    new_data['Date'][x] = data['Date'][x]
    new_data['Close'][x] = data['Close'][x]
```

```python
#Splitting into train and validation
train = new_data[:4218]
valid = new_data[4218:]

x_train = train.drop('Close', axis = 1)
y_train = train['Close']
x_valid = valid.drop('Close', axis = 1)
y_valid = valid['Close']
```
